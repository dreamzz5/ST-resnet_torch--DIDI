```python
import torch

x = torch.tensor([[1., 1.]], requires_grad=True)
c = torch.tensor([[1., 2.]], requires_grad=True)
y = x * c
print(y)
```

```python
#y.backward(torch.ones_like(y))
y[0,0].backward()
print(x.grad)
```
